```python
import numpy as np
from numpy import linalg
import matplotlib.pyplot as plt
import matplotlib

def solve_using_Kaczmarz(A,b):
    print("A=",A);print("b=",b)

    ## number of equations:
    m,n=A.shape
    ## frequency of output during iterations:
    nprint=1
    kmax=6
    
    ## calculate exact solution:
    xsol=linalg.pinv(A)@b
    ## set initial guess at origin:
    Xk=np.zeros((n,1))
    a_i=np.zeros((1,n))
    X=np.zeros((kmax,n))*np.nan
    X[0,:]=Xk.T
    
    print(1,'Iterative solution:')
    k=0
    print('k=%d; X[k]=      ' % (k),end=""); print(Xk)
    for k in range(kmax):
        i=np.mod(k,m)
        a_i=A[i,:]
        a_i.shape=(1,n)
        omega=1
        Xkp1=Xk+omega*(b[i]-a_i@Xk)*a_i.T/linalg.norm(a_i)**2
        if k==np.round(k/nprint)*nprint:
            print('k=%d; i=%d; X[k+1].T=      ' % (k,i),end=""); print(Xkp1.T,end="")
            print(" error=",linalg.norm(xsol-Xkp1))
            Xk=Xkp1
            
            ## save current iteration for plot:
            X[k,:]=Xkp1.T
            
    print('exact solution: ',xsol.T)

    ## plot:
    plt.rcParams['font.family'] = 'Myriad Pro'
    fontsize=12
    plt.rcParams['font.size'] = fontsize
    fig=plt.figure(1,figsize=(4,3))
    plt.clf()
    x=np.arange(-0.2,3.1,0.01)
    y1=-A[0,0]*x/A[0,1]+b[0]/A[0,1]
    y2=-A[1,0]*x/A[1,1]+b[1]/A[1,1]
    plt.plot(x,y1,'b')
    plt.plot(x,y2,'g')
    plt.plot(X[:,0],X[:,1],'--xk',dashes=(20, 10),linewidth=0.25,markersize=3,markeredgewidth=0.5)
    # final iterative solution:
    plt.plot(X[-1,0],X[-1,1],'xr')
    # exact solution:
    plt.plot(xsol[0],xsol[1],'+r')
    plt.axis('equal')
    #plt.ylim(-0.7,0.7)
    plt.grid(lw=0.25)
    plt.xticks(np.arange(0,2.5,0.5))
    plt.xlabel("$x_k$")
    plt.ylabel("$y_k$")
    plt.tight_layout(pad=0)
    plt.pause(0.1)
    #fig.savefig("Output/Kaczmarz.pdf")

    
########################################################################
# Main program:
########################################################################


## set up matrix equation (If you are using this code for HW, replace
## the following two lines with the specified A,b):


# A=np.array([[1.5, -3], [-1, -0.5]]) # 90 degree angle between lines
# solve_using_Kaczmarz(A,b)
# wait=input("enter to continue")

# A=np.array([[1.5, -3], [-1.5, -0.5]]) # larger than 90 degree angle between lines
# solve_using_Kaczmarz(A,b)
# wait=input("enter to continue")
```
